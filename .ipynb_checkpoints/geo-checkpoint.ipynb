```python
'''Импортируем все необходимые библиотеки'''
import json
import random
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
from shapely.geometry import Polygon, Point
from pathlib import Path
import datetime
import ipywidgets as widgets
```

```python
'''1. Написать код для загрузки и визуализации GeoJSON'''
# Загружаем config.json
with open("config.json", "r", encoding="utf-8") as f:
    config = json.load(f)

# Загружаем GeoJSON
gdf = gpd.read_file("area.geojson")

# Переводим в систему координат EPSG:4326 (WGS 84) — долгота/широта
gdf = gdf.set_crs("EPSG:4326")

# Преобразуем в метры (Web Mercator)
gdf = gdf.to_crs("EPSG:3857")

# Проверяем площадь после преобразования
print(f"Площадь зоны застройки после преобразования: {gdf[gdf['name'] == 'Зона застройки'].geometry.iloc[0].area:.2f} м²")

# Визуализация территории
fig, ax = plt.subplots(figsize=(8, 8)) # создаем фигуру и ось
gdf.plot(ax=ax, color="lightpink", edgecolor="gray")
# добавляем подписи к объектам
for idx, row in gdf.iterrows():  
    centroid = row["geometry"].centroid  # находим центр геометрии
    ax.text(centroid.x, centroid.y, row["name"], fontsize=12, ha="center",
            bbox=dict(facecolor="white", alpha=0.7))  # подписываем участок
plt.title("Исходная территория")
plt.show()

```

```python
'''2. Реализовать разбиение территории (жилая, коммерческая, дороги, парки) и определение области застройки'''

min_distance = config["min_distance"] 

# добавляем буферы(область вокруг объекта, которая создаётся на определённом расстоянии)
gdf["buffer"] = gdf.apply(
    lambda row: row.geometry.buffer(min_distance) if row.get("restriction") == "no_build" else row.geometry,
    axis=1
)

# выбираем только объекты с запретом на строительство
buffer_gdf = gdf[gdf["restriction"] == "no_build"][["buffer", "name"]]
buffer_gdf = buffer_gdf.rename(columns={"buffer": "geometry"})  # GeoDataFrame должен содержать "geometry"

# Объединяем все буферные зоны в одну
restricted_area = buffer_gdf.union_all().buffer(0)


# Определяем зону застройки, вычитая из нее буфер запрещенной зоны
buildable_area = gdf[gdf["name"] == "Зона застройки"].geometry.iloc[0].difference(restricted_area)

# Визуализация
fig, ax = plt.subplots(figsize=(8, 8))
gdf.plot(ax=ax, color="gray", edgecolor="black", alpha=0.3)  # Исходные объекты
buffer_gdf.plot(ax=ax, color="lightpink", alpha=0.5, edgecolor="black")  # Запрещённые зоны
gpd.GeoSeries([buildable_area]).plot(ax=ax, color="lightgreen", alpha=0.5, edgecolor="black")  # Разрешённая зона

plt.title("Разделение территории с учётом ограничений")
plt.show()
```

```python
'''3. Алгоритм размещения зданий с учётом ограничений'''
'''Часть 1. Генерируем первое здание случайным образом'''

import random
from shapely.geometry import box, Point
import matplotlib.pyplot as plt

# Функция для создания случайного здания
def generate_building(buildable_area, min_size, max_size):
    """Генерирует случайное прямоугольное здание внутри разрешенной зоны."""
    while True:
        # Выбираем случайную точку внутри `buildable_area`
        minx, miny, maxx, maxy = buildable_area.bounds # НЕ ПОНИМАЮ УТОЧНИТЬ 
        x = random.uniform(minx, maxx)
        y = random.uniform(miny, maxy)
        
        # Случайные размеры здания
        width = random.uniform(min_size, max_size)
        height = random.uniform(min_size, max_size)

        # Создаем прямоугольник здания
        building = box(x, y, x + width, y + height)

        # Проверяем, входит ли здание в зону застройки
        if buildable_area.contains(building):
            return building

# Генерируем одно здание
building = generate_building(buildable_area, config["min_size"], config["max_size"])

# Визуализация
fig, ax = plt.subplots(figsize=(8, 8))
gdf.plot(ax=ax, color="gray", edgecolor="black", alpha=0.3)  # Исходные объекты
gpd.GeoSeries([buildable_area]).plot(ax=ax, color="lightgreen", alpha=0.5, edgecolor="black")  # Разрешённая зона
gpd.GeoSeries([building]).plot(ax=ax, color="blue", alpha=0.7, edgecolor="black")  # Здание

plt.title("Проверка первого здания")
plt.show()
```

```python
'''3. Алгоритм размещения зданий с учётом ограничений'''
'''Часть 2. Генерируем здания, пока не будет достигнут лимит плотности'''
from shapely.ops import unary_union

def generate_buildings(buildable_area, min_size, max_size, min_distance, density_limit):
    buildings = []  # Список зданий
    total_area = buildable_area.area
    built_area = 0  # Общая застроенная площадь

    while built_area / total_area < density_limit:
        # Генерируем новое здание
        new_building = generate_building(buildable_area, min_size, max_size)
         
        # Проверяем расстояние до уже построенных зданий
        if all(new_building.distance(b) >= min_distance for b in buildings):
            buildings.append(new_building)
            built_area += new_building.area  # Увеличиваем застроенную площадь

    return buildings

# Генерация зданий
buildings = generate_buildings(buildable_area, config["min_size"], config["max_size"], config["min_distance"], config["density_limit"])

# Визуализация
fig, ax = plt.subplots(figsize=(8, 8))
gdf.plot(ax=ax, color="gray", edgecolor="black", alpha=0.3)  # Исходные объекты
gpd.GeoSeries([buildable_area]).plot(ax=ax, color="lightgreen", alpha=0.5, edgecolor="black")  # Разрешённая зона
gpd.GeoSeries(buildings).plot(ax=ax, color="blue", alpha=0.7, edgecolor="black")  # Здания

plt.title("Сгенерированные здания с ограничением плотности и расстояния")
plt.show()


```

```python
'''4. Сохранение результата в GeoJSON.'''

# Я решила, что в данном случае наиболее удобным способом записи данных будет подход с уникальными файлами для каждой генерации

# Генерация новых зданий
buildings = generate_buildings(buildable_area, config["min_size"], config["max_size"], config["min_distance"], config["density_limit"])

# Преобразуем здания в GeoSeries
buildings_gdf = gpd.GeoSeries(buildings)

# Создаем имя файла на основе текущего времени
timestamp = datetime.datetime.now().strftime("%d%m%Y_%H%M%S")
file_name = f"buildings_{timestamp}.geojson"

output_path = Path.cwd() / "buildings" / file_name

# Сохраняем данные в файл
# buildings_gdf.to_file(file_name, driver="GeoJSON")
buildings_gdf.to_file(output_path, driver="GeoJSON")

# Определяем путь для сохранения
print(f"Данные сохранены в файл: {file_name}. Файл сохранен в текущую директорию: {output_path}")
```
